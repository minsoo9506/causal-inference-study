# 05 - The Unreasonable Effectiveness of Linear Regression

- online과 offline 수업이 시험결과에 어떤 영향을 주는지 regression 모델을 통해 구한다면 아래와 같이 만들 수 있다. (데이터는 RCT를 통해 구했다고 가정)

$$exam_i = \beta_0 + \kappa \ Online_i + u_i$$

- 이  모델을 이용하여 실제 데이터를 통해 계수를 구하면 우리는 ATE만을 알 수 있는게 아니라 confidence interval, p-value까지 알 수 있다.

- 그렇다면 RCT가 아닌 observational에서는 위와 같은 모델을 만들 수 없을까? 아니다 할 수 있다. 이전 chapter에서 배웠듯이 treatment와 outcome 모두에 영향을 주는 변수들(confounding)을 회귀식의 변수로 넣어서 그 변수가 given된 상태가 됨으로 treatment의 계수를 통해 treatment effect를 구할 수 있다.

```python
import graphviz as gr
```

```python
g = gr.Digraph()
g.edge("IQ", "Educ", color="red"), g.edge("IQ", "Wage", color="red"), g.edge("Educ", "Wage", color="red")
g
```

- 즉, 위에서 $Wage = \beta_0 + \beta_1 Educ + \beta_2 IQ$ 라는 회귀식을 만들어서 $\beta_1$의 값으로 treatment effect를 알 수 있는 것이다. 회귀식에서 $Wage$와 $Educ$의 관계를 고려할 때, $IQ$는 fixed된 상태이므로 아래 그림처럼 다른 변수로 향하는 화살표를 없앨 수 있다. 이렇게 회귀식에서 $IQ$를 고려함으로서 confounding bias를 잡을 수 있게 된다.

```python
g = gr.Digraph()
g.node("IQ=x"), g.edge("Educ", "Wage", color="red")
g
```

- RCT의 경우는 아래와 같다고 할 수 있다.

```python
g = gr.Digraph()
g.edge("IQ", "Wage", color="red"), g.edge("Educ", "Wage", color="red")
g
```
